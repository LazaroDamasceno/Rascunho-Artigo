```python
import polars as pl
```

```python
pl.Config.set_tbl_rows(-1)
pl.Config.set_tbl_cols(-1)
```

```python
df = pl.read_parquet('dados_parquet/')
```

```python
df.shape
```

```python
for col in df.columns:
    has_nulls = df.select(
        pl.col(col).has_nulls().any()
    ).item()
    if (has_nulls):
        print(f'{col} tem nulos')
```

```python
df = df.filter(
   ~pl.col('VINCULACAO_ANO').is_in(['Com dispensa', 'Sem dispensa', 'null'])
).with_columns(
    pl.col('ANO_NASCIMENTO').cast(pl.Int16),
    pl.col('MUN_NASCIMENTO').cast(pl.Categorical),
    pl.col('UF_NASCIMENTO').cast(pl.Categorical),
    pl.col('PAIS_NASCIMENTO').cast(pl.Categorical),
    pl.col('ESTADO_CIVIL').cast(pl.Categorical),
    pl.col('SEXO').cast(pl.Categorical),
    pl.col('ESCOLARIDADE').cast(pl.Categorical),
    pl.col('VINCULACAO_ANO').cast(pl.Int16),
    pl.col('ZONA_RESIDENCIAL').cast(pl.Categorical),
    pl.col('MUN_RESIDENCIA').cast(pl.Categorical),
    pl.col('UF_RESIDENCIA').cast(pl.Categorical),
    pl.col('PAIS_RESIDENCIA').cast(pl.Categorical)
).with_columns(
     convacao = pl.when(
         pl.col('DISPENSA') == 'Sem dispensa'
     ).then(True).otherwise(False)
).drop('DISPENSA')
```
