```python
import polars as pl
import matplotlib.pyplot as plt
```

# ANÁLISE INICIAL

```python
pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_rows(-1)
```

```python
url = 'projecoes_2024_tab1_idade_simples.ods'
```

```python
df = pl.read_ods(url)
```

```python
df.head()
```

```python
df = df.filter(
    pl.col('LOCAL') == 'Brasil'
).drop([
    'SIGLA', 
    'LOCAL',
    'CÓD.'
]).rename({
    'IDADE': 'idade',
    'SEXO': 'sexo'
})
```

```python
df.head()  
```

```python
df_unpivoted =  df.unpivot(
    index=['idade', 'sexo'],
    on=[str(x) for x in range(2000, 2071)],
    variable_name='ano',
    value_name='populacao_projetada'
).with_columns(
    pl.col('idade').cast(pl.Int8),
    pl.col('sexo').cast(pl.Categorical),
    pl.col('ano').cast(pl.Int16),
    pl.col('populacao_projetada').cast(pl.Float32),
)
```

```python
df_unpivoted.head()
```

# ANÁLISE POR ANO E SEXO

```python
df_ano = df_unpivoted.group_by(['ano', 'sexo']).agg(
    pl.col('populacao_projetada').sum()
).sort('ano', descending=False)
```

```python
df_ano.head()
```

```python
populacao_h_max = df_ano.filter(
    pl.col('sexo') ==  'Homens'
).sort(
    'populacao_projetada', 
    descending=True
).row(0)[0]
```

```python
populacao_m_max = df_ano.filter(
    pl.col('sexo') ==  'Mulheres'
).sort(
    'populacao_projetada', 
    descending=True
).row(0)[0]
```

```python
populacao_max = df_ano.filter(
    pl.col('sexo') ==  'Ambos'
).sort(
    'populacao_projetada', 
    descending=True
).row(0)[0]
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

marcadores = ['o', '^', 'd']
anos = sorted(df_ano.select('ano').unique().to_series().to_list())

for (sexo, grupo), marcador in zip(df_ano.group_by('sexo'), marcadores):
    ax.plot(
        grupo.select('ano').to_series(), 
        grupo.select('populacao_projetada').to_series(),
        label=sexo,
        marker=marcador
    )

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.grid()
ax.tick_params('x',  rotation=90)
ax.set_xticks(anos)
ax.set_ylabel('População Projetada')
ax.set_xlabel('Ano')
ax.axvline(2041, color='red', linestyle='--')
ax.axvline(2042, color='red', linestyle='-')

plt.figtext(0.5, -0.03, 
            """Nota: A linha tracejada (2041) indica o pico populacioanl da população brasileira e dos homens.\n Nota: A linha contínua (2042) indica o pico populacional das mulheres.""", 
            ha='center', 
            fontsize=10, 
            style='italic'
)
plt.tight_layout()
fig.savefig('projecao_populacao_ano_sexo.png', dpi=300, bbox_inches='tight')
plt.show()
```
